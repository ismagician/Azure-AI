```python
# Establecemos nuestras credenciales. En un archivo se guarda la key y el endpoint para no exponerlos.
with open('./computer_vision_key.txt') as f:
    lines = [line.rstrip() for line in f]

key = lines[0]
endpoint = lines[1]
```

```python
# Conectamos con Azure

from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from msrest.authentication import CognitiveServicesCredentials

client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(key))
```

```python
# Importamos liberas

import vision
import os

%matplotlib inline

# Descripcin de imagen

image_path = '../img/team.jpg'

image_stream = open(image_path, 'rb')
description = client.describe_image_in_stream(image_stream)

print(description)

vision.show_image_caption(image_path, description)
```

```python
# Descripcin de imagen

image_path = '../img/landscape.jpg'
image_stream = open(image_path, 'rb')
description = client.describe_image_in_stream(image_stream)

vision.show_image_caption(image_path, description)
```

```python
# Descripcin de objetos y carcteristicas

image_path = '../img/team.jpg'

features = ['Description', 'Tags', 'Adult', 'Objects', 'Faces']

image_stream = open(image_path, 'rb')
analysis = client.analyze_image_in_stream(image_stream, visual_features=features)

vision.show_image_analysis(image_path, analysis)
```
